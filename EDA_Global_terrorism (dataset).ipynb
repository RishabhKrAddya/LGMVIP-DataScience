```python
## choosing those columns we have needed and rename it for better understading
df.rename(columns={'iyear':'year','imonth':'month','iday':'day','gname':'group','country_txt':'Country','region_txt':'Region',
                   'attacktype1_txt':'attacktype','targettype1_txt':'targettype','weapontype1_txt':'weapontype','nkill':'kill',
                   'nwound':'wound'},inplace=True)
df.head()
```

#### Basic preliminary analysis of the dataset

```python
print("Country with highest terrorist attack:",df['Country'].value_counts().index[0])
print('Maximum people killed in the attack:',df['kill'].max())
print('Year with the most attack:',df['year'].value_counts().idxmax())
print('Year with the least attack:',df['year'].value_counts().idxmin())
print('Most attack types:',df['attacktype'].value_counts().idxmax())
print('Most active group:',df['group'].value_counts().index[1])
```

#### Countrywise attack

#### Static Visualisation

```python
plt.figure(figsize=(20,7))
p=df['Country'].value_counts().index
sns.countplot(df['Country'],order=p[:26],palette=sns.color_palette('flare'))
plt.xlabel('Country')
plt.ylabel('Death Count')
plt.title('Top 25 countries under threat')
plt.xticks(rotation=90)
plt.show()
```

```python
df['attacktype'].value_counts()
```

```python
plt.rcParams['figure.figsize']=15,11
sns.barplot(x=df['Country'].value_counts()[:20].index,y=df['Country'].value_counts()[:20].values)
plt.ylabel('No. of attacks',)
plt.xticks(rotation=90)
plt.xlabel('Attack Region')
```

```python
plt.figure(figsize=(20,7))
sns.countplot(df['attacktype'],order=df['attacktype'].value_counts().index,palette=sns.color_palette('flare'))
plt.xlabel('Attacking Type')
plt.ylabel('Resulting Death Count')
plt.title('Forms of attack')
plt.xticks(rotation=90)
plt.show()
```

#### Interactive Visualisation

#### Yearwise attack

```python
x_year=df['year'].unique()
y_count_years=df['year'].value_counts(dropna=False).sort_index()
plt.figure(figsize=(12,7))
sns.barplot(x=x_year,y=y_count_years,palette='dark:salmon_r')
plt.xticks(rotation=90)
plt.xlabel('Attack year')
plt.ylabel('No.of attack each year')
plt.title('Attack_of_year')
plt.show()
```

#### Monthwise attacks

```python
plt.rcParams['figure.figsize']=15,11
sns.barplot(x=df['month'].value_counts()[:20].index,y=df['month'].value_counts()[:20].values)
plt.ylabel('No. of  attacks:',fontsize=18)
plt.xticks(rotation=90)
plt.xlabel('Attack month')
plt.title('Attack by month',size=25)
```

#### No. of deaths in different years

```python
plt.figure(figsize=(12,7))
df.groupby(['year'])['kill'].sum().plot(kind='bar',colormap='summer')
plt.title('No. of deaths in different years',fontsize=25)
plt.xlabel('Years',fontsize=25)
plt.ylabel('No. of deaths',fontsize=25)
plt.xticks(rotation=90)
plt.show()
```

#### Bivariate Analysis

```python
plt.subplot(1,2,1)

df.groupby(['Region'])['wound'].sum().sort_values(ascending=False).plot(kind='bar',colormap='rocket')
plt.title('Regions having wounded people',fontsize=15)
plt.xlabel('Region')
plt.ylabel('No of wounded')
plt.xticks(rotation=90)

plt.subplot(1,2,2)
df.groupby(['Region'])['kill'].sum().sort_values(ascending=False).plot(kind='bar',colormap='rocket')
plt.title('Regions having killed people')
plt.xlabel('Region')
plt.ylabel('No of killed')
plt.xticks(rotation=90)
plt.gcf().set_size_inches(15,3)
```

#### Top 10 terrorist attack

```python
plt.figure(figsize=(10,10))
sns.barplot(df['group'].value_counts()[1:11].values,df['group'].value_counts()[1:11].index,palette="rocket")
plt.title('Top 10 terrorist organisation with highest terror attack',fontsize=15)
plt.xlabel('No of attack',fontsize=15)
plt.ylabel('Terrorist Attack',fontsize=15)
plt.show()
```

#### No of deaths over the years

```python
df=df[['year','kill']].groupby(['year']).sum()
fig,ax4=plt.subplots(figsize=(20,20))
df.plot(kind='bar',alpha=0.7,ax=ax4)
plt.xticks(rotation=90)
plt.title('people died due to attack',fontsize=25)
plt.xlabel('year',fontsize=20)
plt.ylabel('No of killed people',fontsize=20)
top_side=ax4.spines['top']
top_side.set_visible(False)
right_side=ax4.spines['right']
right_side.set_visible(False)
```
